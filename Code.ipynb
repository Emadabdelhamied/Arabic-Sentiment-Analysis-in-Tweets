```python
#Postive and negative dataset 
#!/usr/bin/python
#-*-coding :utf8 -*-
from __future__ import unicode_literals
import re
import nltk
file=nltk.data.load("corpora/tweets/traintweets.tsv",format="raw")
dfile=file.decode("utf-8")
print(dfile[0:1000])
```

```python
#segmentation of dataset by english pickle segmenter
def segment(text):
    sens_token=nltk.data.load("tokenizers/punkt/english.pickle")
    sentences=sens_token.tokenize(text)
    text=str(sentences).encode("utf-8")
    dtext=text.decode("utf-8")
    print(dtext)
segment(dfile[0:1000])
```

```python
#tokenize dataset by wordtokenize
def Tokenize ():
    tokens=dfile.split('\t')
    return tokens[0:1000]
def tokenize(text):
    stmts=nltk.word_tokenize(text)
    text=str(stmts).encode("utf-8")
    dtext=text.decode("utf-8")
    print(dtext)
tokenize(dfile[0:1000])
```

```python
def remove_emojis(data):
    emoj = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"\U00002500-\U00002BEF"  # chinese char
        u"\U00002702-\U000027B0"
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        u"\U0001f926-\U0001f937"
        u"\U00010000-\U0010ffff"
        u"\u2640-\u2642" 
        u"\u2600-\u2B55"
        u"\u200d"
        u"\u23cf"
        u"\u23e9"
        u"\u231a"
        u"\ufe0f"  # dingbats
        u"\u3030"
                      "]+", re.UNICODE)
    return re.sub(emoj, '', data)
clean=remove_emojis(s)
print(clean)
```

```python
# removal from dataset
def noramlize(Tweet):
    Tweet = re.sub(r"[ุฅุฃูฑุขุง]", "ุง", Tweet)
    Tweet = re.sub(r"ู", "ู", Tweet)
    Tweet = re.sub(r"ุค", "ุก", Tweet)
    Tweet = re.sub(r"ุฆ", "ุก", Tweet)
    Tweet = re.sub(r'[^ุง-ู ]', "", Tweet)

    noise = re.compile(""" ู    | # Tashdid
                         ู    | # Fatha
                         ู    | # Tanwin Fath
                         ู    | # Damma
                         ู    | # Tanwin Damm
                         ู    | # Kasra
                         ู    | # Tanwin Kasr
                         ู    | # Sukun
                         ู     # Tatwil/Kashida
                     """, re.VERBOSE)
    Tweet = re.sub(noise, '', Tweet)
    return Tweet
noramlize(s)
```
